```python
df['Age2'] = '6'
df.loc[(df['Age'] > 10) & (df['Age'] <= 20), 'Age2'] = '5'
df.loc[(df['Age'] > 20) & (df['Age'] <= 25), 'Age2'] = '4'
df.loc[(df['Age'] > 25) & (df['Age'] <= 30), 'Age2'] = '3'
df.loc[(df['Age'] > 30) & (df['Age'] <= 35), 'Age2'] = '2'
df.loc[(df['Age'] > 35) & (df['Age'] <= 40), 'Age2'] = '1'
df.loc[(df['Age'] > 40) & (df['Age'] <= 45), 'Age2'] = '0'
```

```python
df['New_Family_Income'] = '6'
df.loc[(df['Family_Income'] > 140000) & (df['Family_Income'] <= 200000), 'New_Family_Income'] = '0'
df.loc[(df['Family_Income'] > 200000) & (df['Family_Income'] <= 250000), 'New_Family_Income'] = '1'
df.loc[(df['Family_Income'] > 250000) & (df['Family_Income'] <= 300000), 'New_Family_Income'] = '2'
df.loc[(df['Family_Income'] > 300000) & (df['Family_Income'] <= 350000), 'New_Family_Income'] = '3'
df.loc[(df['Family_Income'] > 350000) & (df['Family_Income'] <= 400000), 'New_Family_Income'] = '4'
df.loc[(df['Family_Income'] > 400000) & (df['Family_Income'] <= 450000), 'New_Family_Income'] = '5'
```
